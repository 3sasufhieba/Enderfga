```python
import numpy as np

#load dirty dataset into a numpy array called data, using numpy.genfromtxt
monthly = np.genfromtxt("Land_and_Ocean_complete.txt",comments='%',filling_values='NaN')
yearly = np.genfromtxt("Land_and_Ocean_summary.txt",comments='%',filling_values='NaN')
```

```python
# 转换为dataframe
import pandas as pd
monthly_df = pd.DataFrame(monthly,columns=['year','month','Monthly','MonthlyUnc.','Annual','AnnualUnc.','Five-year','Five-yearUnc.','Ten-year','Ten-yearUnc.','Twenty-year','Twenty-yearUnc.'] )
yearly_df = pd.DataFrame(yearly,columns=['year','Annual_air','AnnualUnc._air','Five-year_air','Five-yearUnc._air','Annual_water','AnnualUnc._water','Five-year_water','Five-yearUnc._water'] )
```

```python
# 取出Monthly部分的数据，
temp = monthly_df[['year','month','Monthly']]
# 取出数据的一半
df = temp.iloc[0:temp.shape[0]//2,:]
# 生成datetime格式的data列
df['date'] = pd.to_datetime(df[['year','month']].assign(DAY=1))
df = df[['date','Monthly']]
```

```python
import pandas as pd
df = pd.read_csv('2012-2022.csv')
# 前4个数字是年份，后2个数字是月份
df['year'] = df['Year'].apply(lambda x: int(str(x)[:4]))
df['month'] = df['Year'].apply(lambda x: int(str(x)[4:]))
# 将年月合并成一个字段
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str))
# 只保留date和Value字段
df = df[['date','Value']]
```

```python
# 其中增长率最大的是哪个月份
df['pct_change'] = df['Value'].pct_change()
import plotly.express as px
# 绘制在一张图上
fig1 = px.scatter(df, x="date", y="Value", color="pct_change", color_continuous_scale=px.colors.sequential.RdBu_r,size='Value',size_max=20)
fig1.show()
fig2 = px.line(df, x="date", y="pct_change")
fig2.show()
```
